```python
import json
from collections import defaultdict
import re

# Definizione delle keyword da cercare
keywords = ["Paola Egonu", "Vero Volley Monza","Mint Vero Volley Monza","Allianz Vero Volley Milano","Vero Volley Milano","Alessia Orro","Myriam Sylla","Jordan Larson","Marco Gaspari"]

# Funzione per trovare le keyword in un testo
def find_keywords(text):
    found_keywords = []
    for keyword in keywords:
        if re.search(r'\b{}\b'.format(re.escape(keyword)), text, re.IGNORECASE):
            found_keywords.append(keyword)
    return found_keywords

# Lettura dei post e dei commenti dal file JSON
with open('output_with_web.json', 'r') as file:
    data = json.load(file)

# Dizionario per raggruppare i post basati sulle keyword
keyword_posts = defaultdict(list)

# Iterazione sui post e commenti
for item in data:
    text = item.get('content', '')  # Estrai il testo del post o del commento
    found_keywords = find_keywords(text)  # Trova le keyword nel testo
    if found_keywords:  # Se almeno una keyword Ã¨ stata trovata
        for keyword in found_keywords:
            keyword_posts[keyword].append(item)  # Aggiungi il post al dizionario delle keyword trovate

# Scrivi le aggregazioni in un nuovo file JSON
with open('keyword_aggregations.json', 'w') as output_file:
    json.dump(keyword_posts, output_file, indent=4)
```
