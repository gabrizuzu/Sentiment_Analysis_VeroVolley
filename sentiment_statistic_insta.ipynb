```python
import json

with open('data/insta/output.json', 'r') as f:
    data = json.load(f)
```

```python
count_pos = 0
count_neg = 0
count_neu = 0
count_post = 0

for item in data:
    #Let's not consider VeroVolley's posts (as stated in the project specifications)
    if item.get('post_author') != 'verovolley':
        if item.get('sentiment') == 'positive':
            count_pos = count_pos + 1

        if item.get('sentiment') == 'negative':
            count_neg = count_neg + 1

        if item.get('sentiment') == 'neutral':
            count_neu = count_neu + 1
        count_post = count_post + 1

print('Sentiment positive: ', count_pos)
print('Sentiment negative: ', count_neg)
print('Sentiment neutral: ', count_neu)
```

```python
count_post
```

```python
import matplotlib.pyplot as plt

labels = ['Positive', 'Negative', 'Neutral']
values = [count_pos, count_neg, count_neu]

bars = plt.bar(labels, values, color=['green', 'red', 'blue'], width = 0.5)

for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), value, ha='center', va='bottom', fontsize=8, color='black', weight='bold')

plt.xlabel('Type of sentiment', fontsize=10, weight='bold')
plt.ylabel('Values', fontsize=10, weight='bold')
plt.title('SENTIMENTAL ANALYSIS STATISTICS - NEWSPAPER ARTICLES', fontsize=14)

plt.show()
```

```python
count_pos = 0
count_neg = 0
count_neu = 0

count_comments = 0

for item in data:
    comments = item.get('comments', [])
    for comment in comments:
        sentiment = comment.get('sentiment')
        if sentiment == 'positive':
            count_pos = count_pos + 1
        if sentiment == 'negative':
            count_neg = count_neg + 1
        if sentiment == 'neutral':
            count_neu = count_neu + 1
        count_comments = count_comments + 1

print('Sentiment positive: ', count_pos)
print('Sentiment negative: ', count_neg)
print('Sentiment neutral: ', count_neu)
```

```python
import matplotlib.pyplot as plt

labels = ['Positive', 'Negative', 'Neutral']
values = [count_pos, count_neg, count_neu]

bars = plt.bar(labels, values, color=['green', 'red', 'blue'], width = 0.5)

for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), value, ha='center', va='bottom', fontsize=8, color='black', weight='bold')

plt.xlabel('Type of sentiment', fontsize=10, weight='bold')
plt.ylabel('Values', fontsize=10, weight='bold')
plt.title('SENTIMENTAL ANALYSIS STATISTICS - COMMENTS', fontsize=14)

plt.show()
```
