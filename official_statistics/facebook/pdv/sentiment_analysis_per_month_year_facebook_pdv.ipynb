```python
import json

with open('output_fb_pdv.json', 'r') as f:
    data = json.load(f)
```

```python
import re

def check_no_year(date_string):
    # Espressione regolare per controllare se c'Ã¨ un anno nella data
    year_pattern = r'\b\d{4}\b'

    # Cerca il pattern dell'anno nella stringa della data
    match = re.search(year_pattern, date_string)

    # Se non viene trovato l'anno, restituisce True
    return not bool(match)
```

```python
count_pos = [0] * 12
count_neg = [0] * 12
count_neu = [0] * 12

count_pos_2 = [0] * 12
count_neg_2 = [0] * 12
count_neu_2 = [0] * 12

count_post = 0

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for post in data:
    create = post.get('date')
    sentiment = post.get('sentiment')

    count_post += 1

    for i, month in enumerate(months):
        if month in create:
            index = i
            if check_no_year(create):
                count_pos[index] += 1 if sentiment == 'positive' else 0
                count_neg[index] += 1 if sentiment == 'negative' else 0
                count_neu[index] += 1 if sentiment == 'neutral' else 0
            else:
                if '2023' in create:
                    count_pos_2[index] += 1 if sentiment == 'positive' else 0
                    count_neg_2[index] += 1 if sentiment == 'negative' else 0
                    count_neu_2[index] += 1 if sentiment == 'neutral' else 0

print("2024")
print("Positive ", count_pos)
print("Negative ", count_neg)
print("Neutral ", count_neu)

print("")

print("2023")
print("Positive ", count_pos_2)
print("Negative ", count_neg_2)
print("Neutral ", count_neu_2)
```
