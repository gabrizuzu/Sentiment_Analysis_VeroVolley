```python
import json

with open('output_fb_pdv.json', 'r') as f:
    data = json.load(f)
```

```python
import re

def check_no_year(date_string):
    # Espressione regolare per controllare se c'Ã¨ un anno nella data
    year_pattern = r'\b\d{4}\b'

    # Cerca il pattern dell'anno nella stringa della data
    match = re.search(year_pattern, date_string)

    # Se non viene trovato l'anno, restituisce True
    return not bool(match)
```

```python
count_like = [0] * 12
count_like_2 = [0] * 12
count_post = [0] * 12
count_post_2 = [0] * 12

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for post in data:
    create = post.get('date')

    num_likes = post.get('num_likes')

    if ',' in num_likes: num_likes = num_likes.replace(',', '')

    like = float(num_likes)

    for i, month in enumerate(months):
        if month in create:
            index = i
            if check_no_year(create):
                count_like[index] += like
                count_post[index] += 1

            else:
                if '2023' in create:
                    count_like_2[index] += like
                    count_post_2[index] += 1

print("2024")
print("Like ", count_like)
print("Post ", count_post)

print("")

print("2023")
print("Like ", count_like_2)
print("Post ", count_post_2)
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
colors = ['green', 'red', 'blue', 'yellow', 'orange', 'purple', 'pink', 'cyan', 'magenta', 'brown', 'grey', 'black']

bars = plt.bar(months, count_like, color=colors)

for bar, value in zip(bars, count_like):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), value, ha='center', va='bottom', fontsize=8, color='black', weight='bold')

plt.xlabel('Months', fontsize=10, weight='bold')
plt.ylabel('Value of sentiment', fontsize=10, weight='bold')
plt.title('2023 - TOTAL LIKE PER MONTH - POST', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#GRAPH 2: NUMBER OF POST PER MONTH
plt.plot(months, count_post, marker='o', color='red')

# Aggiunta dei valori sopra i punti
for i, txt in enumerate(count_post):
    plt.annotate(txt, (months[i], count_post[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Months', fontsize=10, weight='bold')
plt.ylabel('Number of Posts', fontsize=10, weight='bold')
plt.title('NUMBER OF POSTS PER MONTH', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
count_like = [0] * 12
count_like_2 = [0] * 12
count_comments = [0] * 12
count_comments_2 = [0] * 12

months = ['Jan', 'Febr', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

for post in data:
    comments = post.get('comments', [])

    for comment in comments:
        create = comment.get('date')

        num_likes = comment.get('likes_num')

        if ',' in num_likes: num_likes = num_likes.replace(',', '')

        like = float(num_likes)

        for i, month in enumerate(months):
            if month in create:
                index = i
                if check_no_year(create):
                    count_like[index] += like
                    count_comments[index] += 1
                else:
                    if '2023' in create:
                        count_like_2[index] += like
                        count_comments_2[index] += 1

print("2024")
print("Like ", count_like)
print("Comments ", count_comments)

print("")

print("2023")
print("Like ", count_like_2)
print("Comments ", count_comments_2)
```
