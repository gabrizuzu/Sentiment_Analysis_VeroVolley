```python
import json

with open('output_with_web.json', 'r') as f:
    data = json.load(f)
```

```python
from datetime import datetime

array = []

for post in data:
  # Ottengo la data di pubblicazione del post
  data_pubblicazione = datetime.strptime(post['date'], '%d/%m/%Y')
  mese = data_pubblicazione.month
  anno = data_pubblicazione.year

  array.append({
      'mese': mese,
      'anno': anno,
      'sentiment': post['sentiment'],
      'comments': post['comments']
  })

print(array)
```

```python
count_pos = [0] * 12
count_neg = [0] * 12
count_neu = [0] * 12

count_pos_2 = [0] * 12
count_neg_2 = [0] * 12
count_neu_2 = [0] * 12

count_post = 0

for post in array:
    anno = post['anno']
    mese = post['mese']
    sentiment = post['sentiment']

    if anno == 2024:
        count_list = count_pos if sentiment == 'positive' else (count_neg if sentiment == 'negative' else count_neu)
        count_list[mese - 1] += 1

    elif anno == 2023:
        count_list = count_pos_2 if sentiment == 'positive' else (count_neg_2 if sentiment == 'negative' else count_neu_2)
        count_list[mese - 1] += 1

    count_post += 1

print("2024")
print("Positive ", count_pos)
print("Negative ", count_neg)
print("Neutral ", count_neu)

print("")

print("2023")
print("Positive ", count_pos_2)
print("Negative ", count_neg_2)
print("Neutral ", count_neu_2)
```

```python
count_pos = [0] * 12
count_neg = [0] * 12
count_neu = [0] * 12

count_pos_2 = [0] * 12
count_neg_2 = [0] * 12
count_neu_2 = [0] * 12

count_comments = 0

months = ['gennaio', 'febbraio', 'marzo', 'aprile', 'maggio', 'giugno', 'luglio', 'agosto', 'settembre', 'ottobre', 'novembre', 'dicembre']

for post in array:
    comments = post.get('comments', [])
    for comment in comments:
        count_comments += 1
        create = comment.get('created_at_utc')
        sentiment = comment.get('sentiment')

        for i, month in enumerate(months):
            if month in create:
                index = i
                if '2024' in create:
                    count_pos[index] += 1 if sentiment == 'positive' else 0
                    count_neg[index] += 1 if sentiment == 'negative' else 0
                    count_neu[index] += 1 if sentiment == 'neutral' else 0
                elif '2023' in create:
                    count_pos_2[index] += 1 if sentiment == 'positive' else 0
                    count_neg_2[index] += 1 if sentiment == 'negative' else 0
                    count_neu_2[index] += 1 if sentiment == 'neutral' else 0

print("2024")
print("Positive ", count_pos)
print("Negative ", count_neg)
print("Neutral ", count_neu)

print("")

print("2023")
print("Positive ", count_pos_2)
print("Negative ", count_neg_2)
print("Neutral ", count_neu_2)
```
