```python
import json

with open('output_with_web.json', 'r') as f:
    data = json.load(f)
```

```python
posts = data
posts
```

```python
filtered_data = []
count = 0

for post in posts:
    comments = post['comments']
    filtered_post = {
        'link': post['link'],
        'sentiment': post['sentiment'],
        'comments' : [{'sentiment': comment['sentiment']} for comment in comments]
    }
    filtered_data.append(filtered_post)
    count = count + 1

filtered_data
```

```python
count
```

```python
count_pos = 0
count_neg = 0
count_neu = 0

for item in filtered_data:
    if item.get('sentiment') == 'positive':
        count_pos = count_pos + 1

    if item.get('sentiment') == 'negative':
        count_neg = count_neg + 1

    if item.get('sentiment') == 'neutral':
        count_neu = count_neu + 1

print('Sentiment positive: ', count_pos)
print('Sentiment negative: ', count_neg)
print('Sentiment neutral: ', count_neu)
```
